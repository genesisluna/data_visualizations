```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
mass_shootings = pd.read_csv('Mass Shootings Dataset Ver 5.csv', encoding='latin-1')
```

```python
mass_shootings.head()
```

```python
mass_shootings.dtypes
```

```python
mass_shootings_new = mass_shootings.dropna(axis = 0, how ='any')
```

```python
mass_shootings_new
```

```python
mass_shootings_new['Age'] = mass_shootings_new['Age'].astype(int)
```

```python
mass_shootings_new.dtypes
```

```python
sns.stripplot(x="Mental Health Issues", y="Age", data=mass_shootings_new)
```

```python
ax = sns.barplot(x='Title', y='Total victims', data=mass_shootings_new, palette='twilight_shifted_r')
ax.set_xticklabels(ax.get_xticklabels(), rotation=60, ha="right")
ax.set_title('Number of Victims for Each Shooting')
plt.figure(figsize=(20, 20))
plt.tight_layout()
plt.show()
```

```python
total_victims = mass_shootings_new.groupby('Cause')['Total victims'].sum()
total_victims.plot(kind='pie', title='Victim Deaths by Cause', figsize=(9,9), autopct='%1.2f%%', colormap='tab10')
plt.ylabel('')
plt.tight_layout()
```
